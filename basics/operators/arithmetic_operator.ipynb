```python
# Initializing variables
var1, var2 = 5,2
```

```python
# Print all arithmetic operations
# Addition
print("Addition: ", var1 + var2)
```

```python
# Subtraction
print("Subtraction: ", var1 - var2)
```

```python
# Multiplication
print("Multiplication: ", var1 * var2)
```

```python
# Division
print("Division: ", var1 / var2)
```

```python
# Modulo Division -> Returns remainder
print("Modulo division: ", var1 % var2)
```

```python
# Floor Division -> Returns Quotient
print("Floor division: ", var1 // var2)
```

```python
# Power/Exponent
print("Pow: ", var1 ** var2) # 5 ^ 2
```
